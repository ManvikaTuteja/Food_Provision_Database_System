# **Query 1**

```python
q1 = "select `TimeFrametoPickUp`, count(NumofServing) from `Provide` group by `TimeFrametoPickUp`"
r1 = pd.read_sql(q1,conn)
r1.rename({'count(NumofServing)': 'Servings', 'TimeFrametoPickUp':'Timeframe'}, axis=1, inplace=True)
r1 = r1.sort_values(by='Timeframe',
                     axis=0,     
                     ascending=False,  
                     inplace = False)
r1
```

```python
fig1 = px.line(r1, x="Timeframe", y="Servings",  width = 900, height = 500, title='Number of Servings based on timeframe')
fig1.show()
```

# **Query 2**

```python
q2 ="select `PaymentInfo`, count(*)/53 AS Percentage from `Meal Receiver` group by `PaymentInfo`"
r2 = pd.read_sql(q2,conn)
r2
```

```python
fig2 = px.pie(r2, values='Percentage', names='PaymentInfo', color = 'PaymentInfo',
        color_discrete_map = {'Cash':'royalblue', 'Credit':'gold'}, width = 500, height = 500, title='Payment Preference')
fig2.show()
```

# **Query 3**

```python
q3 ="select `Category`, count(*) from `Meal` group by `Category`"
r3 = pd.read_sql(q3,conn)
r3.rename({'count(*)': 'Count', 'Category':'Food Category'}, axis=1, inplace=True)
r3
```

```python
fig3 = px.bar(r3, y="Food Category", x="Count", orientation='h', title='Food Preference', width = 800, height = 550,text_auto='.2s')
fig3.update_traces(width=0.6)
fig3.update_layout(barmode='stack', yaxis={'categoryorder':'total ascending'})
fig3.show()
```
